```python
# Cell 1
import pandas as pd
```

```python
# Cell 2 — load & parse
df = pd.read_csv('simulados_clean.csv')
df['created_at'] = pd.to_datetime(df['created_at'])
df.head()
```

```python
# Cell 3 — detect clusters
df['cluster_size'] = df.groupby(['aluno_id', 'created_at'])['created_at'].transform('size')
df['cluster_size'].value_counts()
```

```python
# Cell 4 — split
unique_rows = df[df['cluster_size'] == 1].copy()
clustered_rows = df[df['cluster_size'] > 1].copy()
len(unique_rows), len(clustered_rows)
```

```python
# Cell 5 — collapse
areas = ['matematica', 'linguagens', 'humanas', 'natureza']
collapsed = clustered_rows.groupby(['aluno_id', 'created_at'], as_index=False)[areas].mean()
collapsed
```

```python
# Cell 6 — merge & sort
unique_trimmed = unique_rows[['aluno_id', 'created_at'] + areas]
final_df = pd.concat([unique_trimmed, collapsed], ignore_index=True)
final_df = final_df.sort_values(['aluno_id', 'created_at']).reset_index(drop=True)
final_df.head(15)
```

```python
# Cell 7 — drop empties
fully_empty = final_df[areas].isnull().all(axis=1)
final_df = final_df[~fully_empty].reset_index(drop=True)
len(final_df)
```

```python
# Cell 8 — verify & save
final_df.duplicated(subset=['aluno_id', 'created_at'], keep=False).sum()
```

```python
# Cell 9
final_df.to_csv('simulados_paired_ready.csv', index=False)
```

```python
# Cell 10 — pull out one student's full history to inspect, in order
grouped = final_df.groupby('aluno_id')

# pick the first student id that appears in the data
one_student_id = final_df['aluno_id'].iloc[0]

# .get_group() pulls out just that student's rows from the grouped object
one_student = grouped.get_group(one_student_id)
one_student
```

```python
# Cell 11 — build prev/next pairs for just this one student first, to check the logic

one_student = one_student.reset_index(drop=True)  
# reset_index gives this student's rows clean positions 0, 1, 2...
# (without this, .iloc and .loc could disagree, since the original index 
# numbers came from the full 148-row dataframe, not from 0)

pairs = []  # will hold one dictionary per training pair

for i in range(len(one_student) - 1):
    # -1 because the very last test has no "next" test after it —
    # it can only ever be used as a target, never as an input
    
    prev_test = one_student.iloc[i]      # the earlier test in this pair (a single row)
    next_test = one_student.iloc[i + 1]  # the later test in this pair

    pair_row = {
        'aluno_id': one_student.loc[i, 'aluno_id'],
        'prev_matematica': prev_test['matematica'],
        'prev_linguagens': prev_test['linguagens'],
        'prev_humanas': prev_test['humanas'],
        'prev_natureza': prev_test['natureza'],
        'next_matematica': next_test['matematica'],
        'next_linguagens': next_test['linguagens'],
        'next_humanas': next_test['humanas'],
        'next_natureza': next_test['natureza'],
    }
    pairs.append(pair_row)

pairs_df = pd.DataFrame(pairs)
pairs_df
```

```python
# Cell 12 — repeat the pairing logic for every student, not just one

all_pairs = []  # will collect every pair, from every student

for aluno_id, group in final_df.groupby('aluno_id'):
    # iterating directly over a groupby object gives you (key, group) pairs —
    # aluno_id is the student's id, group is that student's rows (same as 
    # what .get_group() gave you in Cell 10, just looped automatically)

    group = group.reset_index(drop=True)  # same reason as before

    for i in range(len(group) - 1):
        prev_test = group.iloc[i]
        next_test = group.iloc[i + 1]

        pair_row = {
            'aluno_id': aluno_id,
            'prev_created_at': prev_test['created_at'],
            'prev_matematica': prev_test['matematica'],
            'prev_linguagens': prev_test['linguagens'],
            'prev_humanas': prev_test['humanas'],
            'prev_natureza': prev_test['natureza'],
            'next_matematica': next_test['matematica'],
            'next_linguagens': next_test['linguagens'],
            'next_humanas': next_test['humanas'],
            'next_natureza': next_test['natureza'],
        }
        all_pairs.append(pair_row)

pairs_df = pd.DataFrame(all_pairs)
pairs_df
```

```python
pairs_df['aluno_id'].nunique()
```

```python
# Cell 13 — find which students have only 1 test (contributed no pairs)

students_in_pairs = pairs_df['aluno_id'].unique()
all_students = final_df['aluno_id'].unique()

missing = [s for s in all_students if s not in students_in_pairs]
print(f"{len(missing)} students with only 1 test: {missing}")

# also useful: see the test count distribution across all students
final_df.groupby('aluno_id').size().value_counts().sort_index()
```

```python
# Cell 14 — save pairs_df as a checkpoint before building features

pairs_df.to_csv('simulados_pairs.csv', index=False)
print(f"Saved {len(pairs_df)} pairs from {pairs_df['aluno_id'].nunique()} students")
```
